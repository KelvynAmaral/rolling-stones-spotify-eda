```python
ultima_decada = rlstones["release_date"].between(pd.to_datetime('2011'), pd.to_datetime('2020'))
```

### Popularidade por album

```python
rls_por_album = rlstones[ultima_decada].groupby("album")["popularity"].sum().sort_values(ascending=False).head(10)

total_por_popularidade = rls_por_album.sum()

df_porcentagem = rls_por_album / total_por_popularidade * 100

labels =  df_porcentagem.index.tolist()
sizes = df_porcentagem.values.tolist()

figura, grafico = plt.subplots(figsize=(18, 6))
grafico.pie(sizes, autopct='%1.1f%%')
grafico.axis('equal')
plt.title("Álbuns mais populares da última década")
plt.legend(labels, loc="best")
plt.show()
```
