```python
df_ultima_decada = df_spotify['release_date'].between(pd.to_datetime('2011'), pd.to_datetime('2021'))
```

```python
df_por_album = df_spotify[df_ultima_decada].groupby("album")["popularity"].mean().sort_values(ascending=False).head(10)
total_popularidade = df_por_album.sum()
df_porcentagem = (df_por_album / total_popularidade) * 100

labels = df_porcentagem.index.tolist()
sizes = df_porcentagem.values.tolist()

figura, grafico = plt.subplots(figsize=(18, 6))

grafico.pie(sizes, autopct='%1.1f%%')
grafico.axis('equal')
plt.title('Álbuns mais populares da última década')
plt.legend(labels, loc='best')
plt.show()
```

```python
def classifica_musica_ao_vivo(df_spotify):
    if df_spotify["liveness"] >= 0.8:
        return True
    else:
        return False
```

```python
df_spotify["ao_vivo"] = df_spotify.apply(classifica_musica_ao_vivo, axis=1)
```
