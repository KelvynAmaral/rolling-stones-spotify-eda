Criando função para classificar musicas ao vivo e gravadas.

```python
def classifica_musica_ao_vivo(df_spotify):
    if df_spotify["liveness"] >= 0.8:
        return True
    else:
        return False
```

```python
df_spotify["ao_vivo"] = df_spotify.apply(classifica_musica_ao_vivo, axis=1)
```

```python
df_spotify.head()
```

```python
df_spotify["ao_vivo"].value_counts()
```

```python
df_gravado_em_estudio = df_spotify[df_spotify["ao_vivo"] == False]
df_ao_vivo = df_spotify[df_spotify["ao_vivo"] == True]
```

### Histograma

```python
df_studio = df_gravado_em_estudio.groupby("album")['loudness'].sum()
df_show = df_ao_vivo.groupby("album")['loudness'].sum()

fig, axes = plt.subplots(1, 2, figsize=(15, 4))

sns.histplot(data=df_studio, bins=20, ax=axes[0])

axes[0].set_title('Soma do barulho das músicas gravadas em estúdio')
axes[0].set_xlabel('Barulho')
axes[0].set_ylabel('Frequencia')

sns.histplot(data=df_show, bins=20, ax=axes[1])

axes[1].set_title('Soma do barulho das músicas ao vivo')
axes[1].set_xlabel('Barulho')
axes[1].set_ylabel('Frequencia')


fig.tight_layout()
plt.show()
```

### Gráfico de densidade

```python
plt.figure(figsize=(10, 5))

sns.kdeplot(data=df_studio, label='Gravado em estúdio', fill=True)
sns.kdeplot(data=df_show, label='Ao vivo', fill=True)


plt.title('Distribuição do baruho das músicas')
plt.xlabel('Barulho')
plt.ylabel('Densidade')
plt.legend()
```

### Teste estatístico
Vamos analisar a distruição dos dados e fazer um teste estastísticos para validar as distrinuições dos dados são normais


```python
from scipy.stats import shapiro
```

```python
stat, p = shapiro(df_studio)
print('Soma do barulho das músicas gravadas em estúdio: ')
print('Estatística= de teste: {:.4f}, p-valor: {:.4f}'.format(stat, p))
if p > 0.05:
    print('A distribuição é normal')
else:
    print('A distribuição não é normal')
```

```python
stat, p = shapiro(df_show)
print('Soma do barulho das músicas ao vivo: ')
print('Estatística= de teste: {:.4f}, p-valor: {:.4f}'.format(stat, p))
if p > 0.05:
    print('A distribuição é normal')
else:
    print('A distribuição não é normal')
```

### Análise de Correlação

Analisar animação da musica

```python
media_por_album = df_spotify.groupby("album")["valence"].mean().reset_index()

media_por_album = media_por_album.rename(columns={"valence": "media_valence"})

media_por_album["sentimento"] = ["positivo" if v >= 0.6 else "negativo" for v in media_por_album["media_valence"]]

media_por_album.groupby("sentimento")["sentimento"].count()
```

```python
media_por_album.head()
```

```python
df_resultado_final = pd.merge(df_spotify, media_por_album, on="album")
df_resultado_final.head(2)
```

```python
matrix_correlacao = df_resultado_final.corr(numeric_only=True)
correlacao_sentimento = matrix_correlacao["media_valence"]
display(correlacao_sentimento)
```

```python
sns.heatmap(correlacao_sentimento.to_frame(), annot=True, cmap='coolwarm')
plt.title('Correlação entre a valência e as variáveis numéricas')
plt.show()
```

```python
sns.scatterplot(x= "media_valence", y="danceability", hue="sentimento", data=df_resultado_final, palette="coolwarm")
plt.xlabel("Valência")
plt.ylabel("Danceabilidade")
plt.title("Valência x Danceabilidade")
plt.show()
```

```python
sns.scatterplot(x= "media_valence", y="liveness", hue="sentimento", data=df_resultado_final, palette="coolwarm")
plt.xlabel("Valência")
plt.ylabel("Liveness")
plt.title("Valência x Liveness")
plt.show()
```
